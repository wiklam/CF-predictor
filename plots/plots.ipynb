```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ac_before = 0.6144284331065194
ac_after = 0.5435593754835577

tc_before = 0.5374706160336614
tc_after = 0.5298806437807785

cf_before = 0.563
cf_after = 0.56

elo_before = 0.678
elo_after = 0.659

labels = ["CodeForces", "AtCoder", "TopCoder", "ELO"]
colors_before = ["yellow", "orangered", "blue", "limegreen"]
colors_after = ["gold", "red", "mediumblue", "forestgreen"]
errs_before = [cf_before, ac_before, tc_before, elo_before]
errs_after = [cf_after, ac_after, tc_after, elo_after]

xs = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
rects = []
mult = 0.9

fig, ax = plt.subplots(figsize=(15, 8))
for err_before, color, x in zip(errs_before, colors_before, xs):
    rects.append(ax.bar([x - width / 2], [err_before], width * mult, color=color))
for err_after, color, x in zip(errs_after, colors_after, xs):
    rects.append(ax.bar([x + width / 2], [err_after], width * mult, color=color))

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel("Error rate", fontsize=15)
ax.set_title("Logarithmic error rates on expected participant position in contest across different rating systems - original vs optimized", fontsize=15)
ax.set_xticks(xs)
ax.set_xticklabels(labels, fontsize=15)

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('%.2f' % height,
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=13)

for rect in rects:
    autolabel(rect)
    
fig.tight_layout()
fig.savefig("parameters.png")
plt.show()
```
