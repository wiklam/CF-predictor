```python
def binary_round(x):
    if x >= 0.5:
        return 1
    return 0
```

```python
def AddUserInfoEntry(user, entry):
    # entry.append(user.contribution)
    # entry.append(user.rating)
    # In this model it is just disturbing 
    #entry.append(user.maxRating)
    return entry

def AddSingleContestEntry(ContestRating, DB, entry):
    #entry.append(ContestRating.rank)
    entry.append(ContestRating.oldRating)
    #entry.append(ContestRating.newRating)
    entry.append(ContestRating.delta())
    contId = ContestRating.contestId
    # entry.append(DB.contests[contId].startTime)
    # Just to debug
    # entry.append(contId)
    return entry

def AddUserContestsEntry(user, time, num, DB, entry):
    contHist = DB.contestHistory[user.nick]
    contLen = len(contHist)
    
    idx = 0
    while idx < contLen:
        contId = contHist[idx].contestId
        contTime = DB.contests[contId].startTime
        if contTime < time:
            break
        idx += 1
    while idx < contLen and num > 0:
        entry = AddSingleContestEntry(contHist[idx], DB, entry)
        idx += 1
        num -= 1
    return entry
    
def CreateUserEntry(user, time, num, DB):
    entry = []
    entry = AddUserInfoEntry(user, entry)
    entry = AddUserContestsEntry(user, time, num, DB, entry)
    return entry
```

```python
NUMOFUSERS = 3000
NUMOFCONTESTS = 15

numOfUsers = 0
X = []
Y = []
for user in DB.users:
    if user in DB.contestHistory:
        if len(DB.contestHistory[user]) > NUMOFCONTESTS:
            Y.append(binary_round(DB.contestHistory[user][0].delta()))
            contId = DB.contestHistory[user][0].contestId
            time = DB.contests[contId].startTime
            entry = CreateUserEntry(DB.users[user], time, NUMOFCONTESTS, DB)
            X.append(entry)
            print(user)
            print(entry)
            print(contId)
            numOfUsers += 1
            if numOfUsers >= NUMOFUSERS:
                break
```

```python
print(len(X))
poly = PolynomialFeatures(2)
var = poly.fit_transform(X)

poly_var_train, poly_var_test, res_train, res_test = train_test_split(var, Y, shuffle = True)

regression = linear_model.LinearRegression()

model = regression.fit(poly_var_train, res_train)
score = model.score(poly_var_test, res_test)
model.predict(poly_var_test)
score
```

```python
def binary_loss_fun(ans, pred):
    diff = np.abs(pred)
    diff = np.array([binary_round(x) for x in diff])
    diff = np.abs(diff - ans)
    diff = [binary_round(x) for x in diff]
    return sum(diff)
n = len(X)
```
